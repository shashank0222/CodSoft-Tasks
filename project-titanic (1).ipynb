## **Data Preprocessing**

### I have missing values in column age by using statistical measures like mean, median, or regression models based on other available features.
### The 'Cabin' column has a large number of missing values. We might choose to drop this column entirely if it's not critical to our analysis.
### For the 'Embarked' column, we could impute missing values with the most frequent value or use advanced techniques like regression models if there is a relationship with other features.

```python
titanic=titanic.drop(columns=['PassengerId', 'Name', 'Cabin', 'Ticket'], axis=1)
titanic.head()
```

```python
### convert the gender to binary 0 and 1
titanic['Sex']=titanic['Sex'].replace({'male':1,'female':0})
titanic.head()
```

```python
# Fill missing values in age column by imputing the median
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
titanic.isna().sum()
```

```python
# Fill missing values in embarked column by imputing the mode
titanic["Embarked"].fillna(titanic["Embarked"].mode()[0], inplace=True)
titanic.isna().sum()
```

```python
titanic['Embarked']=titanic['Embarked'].replace({'S':1,'C':2,'Q':3})
titanic.head()
```

```python
titanic.corr()
```

```python
plt.figure(figsize=(16, 10))
sns.heatmap(titanic.corr(), annot=True)
plt.show()
```

## **define features and target**

```python
x=titanic.drop(columns=['Survived'])
y=titanic['Survived']
```

## **Data scaling**

```python
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
```

## **reshape to y**

```python
y=y.values.reshape(-1,1)
```

## **split data**

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=True,random_state=42)
```

## **Models**

## **1-Logistic Regression**

```python
lg=LogisticRegression(penalty="l2",max_iter=1000,solver="lbfgs")
lg.fit(X_train,y_train)
```

```python
lg.score(X_train,y_train)
```

```python
y_pred=lg.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
